# VK Group Scraper

This group scraper will iterate VK groups beginning at the group number input into 'counter' field. This is for mass data scraping and collection, but can be altered to do individual groups. 

This notebook will be used in tandem with the user/friends notebook. The users gathered from these group's posts can then be iterated against for their full info using the user/friends notebook. User/friend notebook will accept the dataframe output from this notebook. 

```python
#@title  { display-mode: "form" }
#final script
!pip install vk
!pip install tdqm
#@markdown Input your access token below:

import vk
import csv
import pandas as pd 
import time
from tqdm import tqdm
from vk import session

from google.colab import drive
drive.mount('/content/drive')


if __name__ == "__main__": 
    token = "" #@param {type:'string'}
    session = 'access_token=' + token
    vk_api = vk.API(session) 


def get_posts(group_id, offset, count):
  posts = vk_api.wall.get(owner_id=-group_id, offset=offset, count=count, extended=0, v=5.131)['items']
  return posts


def get_posts_count(group_id):
  return vk_api.wall.get(owner_id=-group_id, offset=0, count=1, extended=0, v=5.131)['count']
#@markdown -----



#@markdown Then input the starting group number in the counter field. You can start at one, they begin at one. 
def get_posts_n(group_id, count):

  for i in range(count):
    posts = get_posts(group_id, i * 100, 100)

  posts_id = [post['id'] if 'id' in post else None for post in posts]
  posts_owner_id = [post['owner_id'] if 'owner_id' in post else None for post in posts]
  posts_from_id = [post['from_id'] if 'from_id' in post else None for post in posts]
  posts_created_by = [post['created_by'] if 'created_by' in post else None for post in posts]
  posts_date = [post['date'] if 'date' in post else None for post in posts]
  posts_text = [post['text'] if 'text' in post else None for post in posts]
  posts_reply_post_id = [post['reply_post_id'] if 'reply_post_id' in post else None for post in posts]
  posts_reply_owner_id = [post['reply_owner_id'] if 'reply_owner_id' in post else None for post in posts]
  posts_copyright = [post['copyright'] if 'copyright' in post else None for post in posts]
  posts_likes = [post['likes'] if 'likes' in post else None for post in posts]
  posts_reposts = [post['reposts'] if 'reposts' in post else None for post in posts]
  posts_post_type = [post['post_type'] if 'post_type' in post else None for post in posts]
  posts_attachments = [post['attachments'] if 'attachments' in post else None for post in posts]
  posts_copy_history = [post['copy_history'] if 'copy_history' in post else None for post in posts]
  posts_comments = [post['comments'] if 'comments' in post else None for post in posts]


  return pd.DataFrame(data={'id': posts_id,
                            'owner_id': posts_owner_id,
                            'from_id': posts_from_id,
                            'created_by': posts_created_by,
                            'date': posts_date,
                            'text': posts_text,
                            'reply_post_id': posts_reply_post_id,
                            'reply_owner_id': posts_reply_owner_id,
                            'copyright': posts_copyright,
                            'likes': posts_likes,
                            'reposts': posts_reposts,
                            'post_type': posts_post_type,
                            'attachments': posts_attachments,
                            'copy_history': posts_copy_history,
                            'comments': posts_comments
                            })
  

def get_posts_all(group_id):

  count = get_posts_count(group_id)

  iters = count // 100
  rest = count % 100

  posts = []

  for i in tqdm(range(iters)):
    posts.extend(get_posts(group_id, i * 100, 100))

  posts.extend(get_posts(group_id, iters * 100, rest))


  posts_id = [post['id'] if 'id' in post else None for post in posts]
  posts_owner_id = [post['owner_id'] if 'owner_id' in post else None for post in posts]
  posts_from_id = [post['from_id'] if 'from_id' in post else None for post in posts]
  posts_created_by = [post['created_by'] if 'created_by' in post else None for post in posts]
  posts_date = [post['date'] if 'date' in post else None for post in posts]
  posts_text = [post['text'] if 'text' in post else None for post in posts]
  posts_reply_post_id = [post['reply_post_id'] if 'reply_post_id' in post else None for post in posts]
  posts_reply_owner_id = [post['reply_owner_id'] if 'reply_owner_id' in post else None for post in posts]
  posts_copyright = [post['copyright'] if 'copyright' in post else None for post in posts]
  posts_likes = [post['likes'] if 'likes' in post else None for post in posts]
  posts_reposts = [post['reposts'] if 'reposts' in post else None for post in posts]
  posts_post_type = [post['post_type'] if 'post_type' in post else None for post in posts]
  posts_attachments = [post['attachments'] if 'attachments' in post else None for post in posts]
  posts_copy_history = [post['copy_history'] if 'copy_history' in post else None for post in posts]
  posts_comments = [post['comments'] if 'comments' in post else None for post in posts]


  return pd.DataFrame(data={'id': posts_id,
                            'owner_id': posts_owner_id,
                            'from_id': posts_from_id,
                            'created_by': posts_created_by,
                            'date': posts_date,
                            'text': posts_text,
                            'reply_post_id': posts_reply_post_id,
                            'reply_owner_id': posts_reply_owner_id,
                            'copyright': posts_copyright,
                            'likes': posts_likes,
                            'reposts': posts_reposts,
                            'post_type': posts_post_type,
                            'attachments': posts_attachments,
                            'copy_history': posts_copy_history,
                            'comments': posts_comments
                            })
  

def group_it():
  counter = 1 #@param {type:'integer'}
  group_id = counter 
  while counter < 2200000:
    
    try: 
      get_posts_count(group_id)
      posts_big_df = get_posts_all(group_id)
      len(posts_big_df)
      %cd /content/drive/MyDrive/vkmining
      filename = 'posts_big_df_' + str(group_id) + '.csv'
      posts_big_df.to_csv(filename)
    except: 
      print('error, restarting. Group ID on failure: ' + str(group_id) )
    counter += 1
    group_id += 1
    import time
    time.sleep(20)
print('Script Running')
group_it()
#@markdown Press Play
```
